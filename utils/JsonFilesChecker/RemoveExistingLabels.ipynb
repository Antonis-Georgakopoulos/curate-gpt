```python
import json
import os

# Load the evaluation dataset
with open('C:\\Users\\antge\\Desktop\\Thesis\\CurateGPT-forked\\curate-gpt\\data\\structured\\BAO_eval_set.json', 'r') as file:
    evaluation_data = json.load(file)

# Extract prefLabels and ids from the evaluation dataset
eval_prefLabels = {entry['prefLabel'] for entry in evaluation_data if 'prefLabel' in entry}
eval_ids = {entry['id'] for entry in evaluation_data}

# Folder containing the second dataset JSON files
folder_path = '.\\'


# Function to process each JSON file
def process_file(file_path):
    with open(file_path, 'r') as file:
        data = json.load(file)
    
    # Filter entities based on prefLabel and id
    filtered_data = []
    for entity in data:
        if 'prefLabel' in entity and entity['prefLabel'] in eval_prefLabels:
            continue
        if entity['id'] in eval_ids:
            continue
        print(file_path)
        # Filter relationships based on target
        entity['relationships'] = [
            rel for rel in entity['relationships']
            if rel['target'] not in eval_prefLabels
        ]
        filtered_data.append(entity)
    
    # Save the filtered data back to the file
    with open(file_path, 'w') as file:
        json.dump(filtered_data, file, indent=4)

# Process all JSON files in the folder
for filename in os.listdir(folder_path):
    if filename.endswith('.json'):
        file_path = os.path.join(folder_path, filename)
        process_file(file_path)

print("Processing complete.")
```

```python
import os
import json

def is_valid_entity(entity):
    """Check if the entity contains id, prefLabel, and relationships."""
    required_keys = ['id', 'label', 'relationships']
    return all(key in entity and entity[key] for key in required_keys)


def process_json_files():
    """Process all JSON files in the current directory and save valid entities to new files."""
    current_directory = os.getcwd()
    output_prefix = "output_"

    # Iterate over all files in the current directory
    for filename in os.listdir(current_directory):
        if filename.endswith(".json") and not filename.startswith(output_prefix):
            input_filepath = os.path.join(current_directory, filename)
            
            with open(input_filepath, 'r') as file:
                entities = json.load(file)
                
                # Filter valid entities
                valid_entities = [entity for entity in entities if is_valid_entity(entity)]
                
                # Construct the output filename
                output_filename = f"{output_prefix}{filename}"
                output_filepath = os.path.join(current_directory, output_filename)
                
                # Save the valid entities to the output file
                with open(output_filepath, 'w') as outfile:
                    json.dump(valid_entities, outfile, indent=4)

# Process the JSON files and save the valid entities
process_json_files()
```
